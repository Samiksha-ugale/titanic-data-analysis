```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

```python
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
gender_df = pd.read_csv("gender_submission.csv")
```

```python
train_df['source'] = 'train'
test_df['source'] = 'test'
test_df['Survived'] = np.nan
combined = pd.concat([train_df, test_df], ignore_index=True)
```

```python
# Only run this if 'Name' is still present:
if 'Name' in combined.columns:
    # Extract title
    combined['Title'] = combined['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)

    # Clean titles
    combined['Title'] = combined['Title'].replace(['Mlle', 'Ms', 'Mme', 'Lady', 'Countess', 'Dona'], 'Miss')
    combined['Title'] = combined['Title'].replace(['Don', 'Major', 'Capt', 'Jonkheer', 'Rev', 'Col', 'Dr', 'Sir'], 'Rare')

    # Now drop 'Name'
    combined.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
else:
    print("'Name' column is already dropped.")
```

```python
train_df.info()
train_df.describe(include='all')
train_df.isnull().sum()

# Advanced: Visualize null values
plt.figure(figsize=(10,5))
sns.heatmap(train_df.isnull(), cmap='viridis', cbar=False, yticklabels=False)
plt.title("Missing Values in Train Data")
plt.show()

# Value counts
print(train_df['Pclass'].value_counts())
print(train_df['Sex'].value_counts())
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(combined.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap - Combined Data")
plt.show()
```

```python
missing = combined.isnull().mean().round(3) * 100
missing = missing[missing > 0]
print("Missing value percentages:")
print(missing.sort_values(ascending=False))
```

```python
# Step 1: Create FamilySize
train_df['FamilySize'] = train_df['SibSp'] + train_df['Parch'] + 1

# Step 2: Create IsAlone based on FamilySize
train_df['IsAlone'] = 0
train_df.loc[train_df['FamilySize'] == 1, 'IsAlone'] = 1
```

```python
# Extract Title from Name
train_df['Title'] = train_df['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)

# Clean up rare and similar titles
train_df['Title'] = train_df['Title'].replace(
    ['Mlle', 'Ms', 'Mme', 'Lady', 'Countess', 'Dona'], 'Miss')
train_df['Title'] = train_df['Title'].replace(
    ['Capt', 'Don', 'Major', 'Col', 'Sir', 'Jonkheer', 'Dr', 'Rev'], 'Rare')
```

```python
pivot = train_df.pivot_table(index='Title', columns='Pclass', values='Survived', aggfunc='mean')
sns.heatmap(pivot, annot=True, cmap='YlGnBu')
plt.title("Survival Rate by Title and Pclass")
plt.show()
```
